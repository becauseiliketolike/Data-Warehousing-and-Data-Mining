**Note:** A Pandas Series is like a column in a table.
It is a one-dimensional array holding data of any type.

```python
a = [1, 7, 2]

series_dataset = pd.Series(a)

print(series_dataset)
```

Custom Labels

```python
a = [1, 7, 2]

custom_label = pd.Series(a, index = ["a", "b","c"])

print(custom_label)
```

From Json like key/value object pairs

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

calories_df = pd.Series(calories)

print(calories_df)
```

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)
```

```python
print(df.loc[2])
```

```python
print(df.loc[[1,2]])
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df)
```

```python
print(df.loc["day2"])
```
